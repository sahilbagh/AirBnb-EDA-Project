#### Chart - 1 What are the Room types and their Prices according to particular area ?

````python
room_price_area_wise = airbnb_dataset.groupby(['neighbourhood_group','room_type'])['price'].max().reset_index()
room_price_area_wise.sort_values(by='price',ascending=False).head(10)
````

````python
#visualization code
neighbourhood_group = ['Brooklyn','Manhattan','Queens','Manhattan','Brooklyn','Stateb Island','Queens','Bronx','Queens','Bronx']
room_type = ['Entire home/apt','Entire home/apt','Private room','Private room','Private room','Entire home/apt','Entire home/apt','Private room','Shared room','Entire home/apt']
#create a dictionary named room_dict to store the count of each room type
room_dict = {}
for i in room_type:
  room_dict[i]=room_dict.get(i,0)+1

plt.bar(room_dict.keys(),room_dict.values(),color='orange',edgecolor='blue')
plt.title('Room Types')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()

````

##### 1. Why did you pick the specific chart?

A bar chart is used when we want to show a distribution of data points or perform a comparison of metric values across different subgroups of  data

##### 2. What is/are the insight(s) found from the chart?

We find that the **Entire home/apt** is the highest number of room types overall and prices are high in the **Brooklyn** and **Manhattan**

````python
area_reviews = airbnb_dataset.groupby(['neighbourhood_group'])['number_of_reviews'].max().reset_index()
area_reviews
````

#### **Chart - 3 What can we learn from the Data? (ex:price)**

````python
price_area = airbnb_dataset.groupby(['price'])['number_of_reviews'].max().reset_index()
price_area
````

````python
#visualization code
price_list = price_area['price']
review = price_area['number_of_reviews']
fig = plt.figure(figsize=(10,5))

plt.scatter(price_list,review)
plt.title('Number of reviews vs Price')
plt.xlabel('Price')
plt.ylabel('Number of reviews')
plt.show()
````

##### 1. Why did you pick the specific chart?

Scatter plots' is used to observe and show relationships between two numeric variables. The dots in a scatter plot not only report the values of individual data points, but also patterns when the data are taken as a whole

##### 2. What is/are the insight(s) found from the chart?

From the above visualization we can say that most number of people like to stay in less price and their review are higher in those areas
